**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 1
# Classe principal de presen√ßa de alunos
class PresencaAluno:
    def __init__(self, nome, matricula, turma):
        self.nome = nome                # Nome do aluno
        self.matricula = matricula      # Matr√≠cula do aluno
        self.turma = turma              # Turma do aluno
        self.presencas = 0              # Contador de presen√ßas

    def registrar_presenca(self):
        self.presencas += 1             # Adiciona uma presen√ßa

    def exibir_dados(self):
        print(f"Nome: {self.nome}")
        print(f"Matr√≠cula: {self.matricula}")
        print(f"Turma: {self.turma}")
        print(f"Presen√ßas: {self.presencas}")
        print("------------------------")

# Criando objetos da classe
aluno1 = PresencaAluno("Lucas Silva", "A123", "3A")
aluno2 = PresencaAluno("Mariana Costa", "B456", "3B")

# Registrando presen√ßa
aluno1.registrar_presenca()
aluno1.registrar_presenca()
aluno2.registrar_presenca()

# Exibindo os dados dos alunos
aluno1.exibir_dados()
aluno2.exibir_dados()
```

## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 2
# Classe principal
class PresencaAluno:
    def __init__(self, nome, matricula, turma):
        self.nome = nome
        self.matricula = matricula
        self.turma = turma
        self.presencas = 0
        self.faltas = 0

    def registrar_presenca(self):
        self.presencas += 1

    def faltou(self):
        self.faltas += 1

    def exibir_dados(self):
        print(f"Nome: {self.nome}")
        print(f"Matr√≠cula: {self.matricula}")
        print(f"Turma: {self.turma}")
        print(f"Presen√ßas: {self.presencas}")
        print(f"Faltas: {self.faltas}")
        print("------------------------")


# Subclasse
class AlunoMonitor(PresencaAluno):
    def __init__(self, nome, matricula, turma, disciplina_monitoria):
        super().__init__(nome, matricula, turma)  # Chamada ao construtor da classe pai
        self.disciplina_monitoria = disciplina_monitoria

    def ajudar_colegas(self):
        print(f"{self.nome} est√° ajudando colegas na disciplina de {self.disciplina_monitoria}.")

    def exibir_dados(self):
        super().exibir_dados()
        print(f"Disciplina de Monitoria: {self.disciplina_monitoria}")
        print("------------------------")


# Testando a classe e a subclasse

# Objeto da classe principal
aluno1 = PresencaAluno("Carlos Eduardo", "C789", "3C")
aluno1.registrar_presenca()
aluno1.faltou()
aluno1.exibir_dados()

# Objeto da subclasse
monitor1 = AlunoMonitor("Julia Lima", "D012", "3C", "Matem√°tica")
monitor1.registrar_presenca()
monitor1.registrar_presenca()
monitor1.ajudar_colegas()
monitor1.exibir_dados()
```

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 3
class PresencaAluno:
    def __init__(self, nome, matricula, turma):
        self.nome = nome
        self.matricula = matricula
        self.turma = turma
        self.__presencas = 0  # Atributo privado
        self.faltas = 0

    # M√©todo get
    def get_presencas(self):
        return self.__presencas

    # M√©todo set
    def set_presencas(self, quantidade):
        if quantidade >= 0:
            self.__presencas = quantidade
        else:
            print("Quantidade inv√°lida de presen√ßas!")

    def registrar_presenca(self):
        self.__presencas += 1

    def faltou(self):
        self.faltas += 1

    def calcular_percentual_presenca(self, total_aulas):
        if total_aulas == 0:
            return 0
        return (self.__presencas / total_aulas) * 100

    def exibir_dados(self):
        print(f"Nome: {self.nome}")
        print(f"Matr√≠cula: {self.matricula}")
        print(f"Turma: {self.turma}")
        print(f"Presen√ßas: {self.__presencas}")
        print(f"Faltas: {self.faltas}")
        print("------------------------")


# Testando o encapsulamento e a l√≥gica
aluno1 = PresencaAluno("Beatriz Souza", "E345", "3D")

# Usando os m√©todos p√∫blicos
aluno1.registrar_presenca()
aluno1.registrar_presenca()
aluno1.faltou()

aluno1.exibir_dados()

# Usando o m√©todo com l√≥gica interna
percentual = aluno1.calcular_percentual_presenca(total_aulas=5)
print(f"Percentual de presen√ßa de {aluno1.nome}: {percentual:.2f}%")

# Testando set com valor inv√°lido
aluno1.set_presencas(-2)  # Deve exibir mensagem de erro

# Alterando o valor com set
aluno1.set_presencas(4)
print(f"Novas presen√ßas: {aluno1.get_presencas()}")
```
