```python
# Desenvolva aqui o cÃ³digo da Parte 2
# Classe principal
class PresencaAluno:
    def __init__(self, nome, matricula, turma):
        self.nome = nome
        self.matricula = matricula
        self.turma = turma
        self.presencas = 0
        self.faltas = 0

    def registrar_presenca(self):
        self.presencas += 1

    def faltou(self):
        self.faltas += 1

    def exibir_dados(self):
        print(f"Nome: {self.nome}")
        print(f"MatrÃ­cula: {self.matricula}")
        print(f"Turma: {self.turma}")
        print(f"PresenÃ§as: {self.presencas}")
        print(f"Faltas: {self.faltas}")
        print("------------------------")


# Subclasse
class AlunoMonitor(PresencaAluno):
    def __init__(self, nome, matricula, turma, disciplina_monitoria):
        super().__init__(nome, matricula, turma)  # Chamada ao construtor da classe pai
        self.disciplina_monitoria = disciplina_monitoria

    def ajudar_colegas(self):
        print(f"{self.nome} estÃ¡ ajudando colegas na disciplina de {self.disciplina_monitoria}.")

    def exibir_dados(self):
        super().exibir_dados()
        print(f"Disciplina de Monitoria: {self.disciplina_monitoria}")
        print("------------------------")


# Testando a classe e a subclasse

# Objeto da classe principal
aluno1 = PresencaAluno("Carlos Eduardo", "C789", "3C")
aluno1.registrar_presenca()
aluno1.faltou()
aluno1.exibir_dados()

# Objeto da subclasse
monitor1 = AlunoMonitor("Julia Lima", "D012", "3C", "MatemÃ¡tica")
monitor1.registrar_presenca()
monitor1.registrar_presenca()
monitor1.ajudar_colegas()
monitor1.exibir_dados()
```

**CÃ³digo:**

```python
# Desenvolva aqui o cÃ³digo da Parte 3
class PresencaAluno:
    def __init__(self, nome, matricula, turma):
        self.nome = nome
        self.matricula = matricula
        self.turma = turma
        self.__presencas = 0  # Atributo privado
        self.faltas = 0

    # MÃ©todo get
    def get_presencas(self):
        return self.__presencas

    # MÃ©todo set
    def set_presencas(self, quantidade):
        if quantidade >= 0:
            self.__presencas = quantidade
        else:
            print("Quantidade invÃ¡lida de presenÃ§as!")

    def registrar_presenca(self):
        self.__presencas += 1

    def faltou(self):
        self.faltas += 1

    def calcular_percentual_presenca(self, total_aulas):
        if total_aulas == 0:
            return 0
        return (self.__presencas / total_aulas) * 100

    def exibir_dados(self):
        print(f"Nome: {self.nome}")
        print(f"MatrÃ­cula: {self.matricula}")
        print(f"Turma: {self.turma}")
        print(f"PresenÃ§as: {self.__presencas}")
        print(f"Faltas: {self.faltas}")
        print("------------------------")


# Testando o encapsulamento e a lÃ³gica
aluno1 = PresencaAluno("Beatriz Souza", "E345", "3D")

# Usando os mÃ©todos pÃºblicos
aluno1.registrar_presenca()
aluno1.registrar_presenca()
aluno1.faltou()

aluno1.exibir_dados()

# Usando o mÃ©todo com lÃ³gica interna
percentual = aluno1.calcular_percentual_presenca(total_aulas=5)
print(f"Percentual de presenÃ§a de {aluno1.nome}: {percentual:.2f}%")

# Testando set com valor invÃ¡lido
aluno1.set_presencas(-2)  # Deve exibir mensagem de erro

# Alterando o valor com set
aluno1.set_presencas(4)
print(f"Novas presenÃ§as: {aluno1.get_presencas()}")
```
